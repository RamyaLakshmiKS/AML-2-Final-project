# Read Kaggle API Key from kaggle.json
Locate the kaggle.json file and read the username and key to set up environment variables.

```python
import json
import os

with open('../kaggle.json') as f:
    creds = json.load(f)

os.environ['KAGGLE_USERNAME'] = creds['username']
os.environ['KAGGLE_KEY'] = creds['key']
```

# Authenticate with Kaggle
Use the Kaggle API to authenticate using the credentials from kaggle.json.

```python
from kaggle.api.kaggle_api_extended import KaggleApi

api = KaggleApi()
api.authenticate()
```

# Download Dataset to Data Folder
Download a specified dataset from Kaggle and save it inside a 'data' folder.

```python
dataset = 'resplan/resplan'
file_name = 'ResPlan.pkl'
api.dataset_download_file(dataset, file_name, path='../data/', force=True)
```
